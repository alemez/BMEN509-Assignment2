```python
#Andrea Lemez
#Assignment 2 Question 2

import matplotlib.pyplot as plt
import numpy as np
```

```python
#f(x, y) = 20 sin(2π ∗ f0x)


Img1 = np.zeros((15,15))
fo1 = 1
ImgSize1 = Img1.shape[0]*Img1.shape[1]


for x in range(Img1.shape[0]):
    for y in range(Img1.shape[1]):
        Img1[x,y] = 20*np.sin(2*np.pi*fo1*x)

freqArr1 = np.array(Img1)
ImgFFT1 = np.fft.fft(freqArr1)
ImgFFT1 = abs(ImgFFT1)*2/ImgSize1
ImgFFTMagShift = np.fft.fftshift(ImgFFT1)        
        
plt.subplots(1, 3, figsize=(20,10))
plt.subplot(1, 3, 1); plt.imshow(Img1, cmap='gray'); plt.title('Signal')
plt.subplot(1, 3, 2); plt.hist(Img1.ravel(), bins=256); plt.title('Histogram')
plt.subplot(1, 3, 3); plt.plot(ImgFFTMagShift); plt.title('Frequency Spectrum')
plt.show()
```

```python
#f(x, y) = 20 sin(2π ∗ f0x) + 20 sin(2π ∗ f0y) 
 

Img2 = np.ones((15,15))
fo2 = 3
ImgSize2 = Img2.shape[0]*Img2.shape[1]


for x in range(Img2.shape[0]):
    for y in range(Img2.shape[1]):
        Img2[x, y] = 20*np.sin(2*np.pi*fo2*x) + 20*np.sin(2*np.pi*fo2*y)

freqArr2 = np.array(Img2)
ImgFFT2 = np.fft.fft(freqArr2)
ImgFFT2M = abs(ImgFFT2)*2/ImgSize2*2/ImgSize2
        
plt.subplots(1, 3, figsize=(20,10))
plt.subplot(1, 3, 1); plt.imshow(Img2, cmap='gray'); plt.title('Signal')
plt.subplot(1, 3, 2); plt.hist(Img2.ravel(), bins=256); plt.title('Histogram')
plt.subplot(1, 3, 3); plt.plot(ImgFFT2M); plt.title('Frequency Spectrum')
plt.show()

```

```python
#f(x, y) = 1 −10 < x < 10 − 10 < y < 10
#          0 otherwise

Img3 = np.ones((15,15))
fo3 = 3
ImgSize3 = Img3.shape[0]*Img3.shape[1]


for x in range(Img3.shape[0]):
    for y in range(Img3.shape[1]):
        if (-10 < x < 10 and -10 < y < 10):
            Img3[x,y] = 1
        else:
            Img3[x,y] = 0
        

freqArr3 = np.array(Img3)
ImgFFT3 = np.fft.fft(freqArr3)
ImgFFT3M = abs(ImgFFT3)*2/ImgSize3*2/ImgSize3        
        
plt.subplots(1, 3, figsize=(20,10))
plt.subplot(1, 3, 1); plt.imshow(Img3, cmap='gray'); plt.title('Signal')
plt.subplot(1, 3, 2); plt.hist(Img3.ravel(), bins=256); plt.title('Histogram')
plt.subplot(1, 3, 3); plt.plot(ImgFFT3M); plt.title('Frequency Spectrum')
plt.show()
```
